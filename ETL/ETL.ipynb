# HOTENCODING

Es necesario convertir los campos categóricos a numéricos pues los algoritmos de Machine Learning requieren valores numéricos para operar.

```python
df_parquet= '/content/drive/MyDrive/ANALISIS DE CARRITO ABANDONADO/1- EDA/Resultado/df_unido_procesado.parquet'
```

```python
df=pd.read_parquet(df_parquet)
```

```python
df
```

```python
df.info()
```

# Convertimos  Product Category a valores numericos

```python
df_category_discretizado=pd.get_dummies(df, columns=['Product Category'])
```

```python
df_category_discretizado
```

# Convertimos Payment Method a Valores numericos

```python
df_category_discretizado1=pd.get_dummies(df_category_discretizado, columns=['Payment Method'])
```

```python
df_category_discretizado1
```

```python
df_category_discretizado1.info()
```

```python
df_category_discretizado2=pd.get_dummies(df_category_discretizado1, columns=['Gender'])
```

```python
df_category_discretizado2
```

```python
df_category_discretizado2 = df_category_discretizado2.rename(columns={'Gender_Female': 'Female', 'Gender_Male': 'Male'})
```

```python
df_category_discretizado2
```

# Extraemos Año, Mes, Dia , Hora en Purchase Date

```python
import datetime as dt
#df_category_discretizado2['Year']=df_category_discretizado2['Purchase Date'].dt.year
df_date= pd.to_datetime(df_category_discretizado2['Purchase Date'])
```

```python
df_category_discretizado2['Year']=df_date.dt.year
```

```python
df_category_discretizado2
```

# Purchase Date Mes

```python
df_category_discretizado2['Month']=df_date.dt.month
```

```python
df_category_discretizado2
```

```python
df_category_discretizado2['Day']=df_date.dt.day
```

```python
df_category_discretizado2
```

```python
df_category_discretizado2['Hour']=df_date.dt.hour
```

```python
df_category_discretizado2
```
